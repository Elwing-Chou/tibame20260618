```python
def is_leap_year(year):
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False

print(is_leap_year(2000))
print(is_leap_year(1800))
```

```python
def is_date_valid(d):
    # 把年月日分開
    day = d % 100
    remain = d // 100
    month = remain % 100
    year = remain // 100
    # print(year, month, day)

    if year >= 1:
        if 1 <= month <= 12:
            max_days_per_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            max_days = max_days_per_month[month-1]
            # if month == 2 and is_leap_year(year) == True
            if month == 2 and is_leap_year(year):
                max_days = max_days + 1
            if 1 <= day <= max_days:
                return True
            else:
                # day invalid
                return False
        else:
            # month invalid
            return False
    else:
        # year invalid
        return False

print(is_date_valid(20000229))
print(is_date_valid(20230229))
```
