```python
# list初始化
# 1. 都一樣
l = [0] * 5
print(l)
# 2. 裡面不一樣
l = []
for i in range(5):
    l.append(i)
print(l)
# 3. 是2的簡化版
l = [(i+2) for i in range(5)]
print(l)
```

```python
# 二維list初始化
# 錯誤方法: 因為你其實a只創造了一次
# 這個也是錯的喔 l = [[0] * 2] * 5
a = [0, 0]
l = [a] * 5
print(l)
l[0][0] = 100
print(l)
# 正確做法 2(跟上面1d的作法2).
l = []
for i in range(5):
    a = [i] * 2
    l.append(a)
l[0][0] = 100
print(l)
# 正確做法 3
l = [[i] * 2 for i in range(5)]
l[0][0] = 100
print(l)
```
