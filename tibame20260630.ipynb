群集型態

1. list: 清單(同類型東西的組合) [20, 30, 40] key=0 1 2 3 有序

2. dict: 字典(拿來聚集成一個複雜東西) {"姓名":"Elwing", "身高":175} key=自訂(數字/字串) 無序

3. set: 不重複的集合(使用上比較接近list) {20, 30, 10} key你無法決定 無序

4. tuple: 有共識版本的字典 (200, 200, 200) key=0 1 2 3 有序  沒有太多操作, 你甚至無法增加東西到tuple裡

```python
s = {20, 30, 10, 30}
print(s)
```

```python
# (|)聯集: 兩個set有的東西都列出來(所有的)
# (&)交集: 兩個set都要有才列出來(重複的)
# (-)減集: A - B A扣掉B = A這個set - AB重複的(交集)
s1 = {20, 30, 40}
s2 = {30, 40, 50}
print(s1 | s2)
print(s1 & s2)
print(s1 - s2)

for c in s1:
    print("!", c)

import random
# 如果你用list
lottery = []
while len(lottery) < 6:
    n = random.randint(1, 49)
    if not n in lottery:
        lottery.append(n)
print(lottery)

# 但如果你用set
lottery = set()
while len(lottery) < 6:
    n = random.randint(1, 49)
    # 加入set: add
    lottery.add(n)
print(lottery)
```
